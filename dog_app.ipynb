## Step 1: Detect Humans

We use OpenCV's implementation of [Haar feature-based cascade classifiers](http://docs.opencv.org/trunk/d7/d8b/tutorial_py_face_detection.html) to detect human faces in images.  OpenCV provides many pre-trained face detectors, stored as XML files on [github](https://github.com/opencv/opencv/tree/master/data/haarcascades).  We have downloaded one of these detectors and stored it in the `haarcascades` directory.

In the next code cell, we demonstrate how to use this detector to find human faces in a sample image.

```python
import cv2                
import matplotlib.pyplot as plt                        
%matplotlib inline                               

# extract pre-trained face detector
face_cascade = cv2.CascadeClassifier('haarcascades/haarcascade_frontalface_alt.xml')

# load color (BGR) image
img = cv2.imread(human_files[3])
# convert BGR image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# find faces in image
faces = face_cascade.detectMultiScale(gray)

# print number of faces detected in the image
print('Number of faces detected:', len(faces))

# get bounding box for each detected face
for (x,y,w,h) in faces:
    # add bounding box to color image
    cv2.rectangle(img,(x,y),(x+w,y+h),(255,0,0),2)
    
# convert BGR image to RGB for plotting
cv_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# display the image, along with bounding box
plt.imshow(cv_rgb)
plt.show()
```

Before using any of the face detectors, it is standard procedure to convert the images to grayscale.  The `detectMultiScale` function executes the classifier stored in `face_cascade` and takes the grayscale image as a parameter.  

In the above code, `faces` is a numpy array of detected faces, where each row corresponds to a detected face.  Each detected face is a 1D array with four entries that specifies the bounding box of the detected face.  The first two entries in the array (extracted in the above code as `x` and `y`) specify the horizontal and vertical positions of the top left corner of the bounding box.  The last two entries in the array (extracted here as `w` and `h`) specify the width and height of the box.

### Write a Human Face Detector

We can use this procedure to write a function that returns `True` if a human face is detected in an image and `False` otherwise.  This function, aptly named `face_detector`, takes a string-valued file path to an image as input and appears in the code block below.

```python
# returns "True" if face is detected in image stored at img_path
def face_detector(img_path):
    img = cv2.imread(img_path)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = face_cascade.detectMultiScale(gray)
    return len(faces) > 0
```

```python
from keras.applications.resnet50 import ResNet50

# define ResNet50 model
ResNet50_model = ResNet50(weights='imagenet')
```

### Pre-process the Data

When using TensorFlow as backend, Keras CNNs require a 4D array (which we'll also refer to as a 4D tensor) as input, with shape

$$
(\text{nb_samples}, \text{rows}, \text{columns}, \text{channels}),
$$

where `nb_samples` corresponds to the total number of images (or samples), and `rows`, `columns`, and `channels` correspond to the number of rows, columns, and channels for each image, respectively.  

The `path_to_tensor` function below takes a string-valued file path to a color image as input and returns a 4D tensor suitable for supplying to a Keras CNN.  The function first loads the image and resizes it to a square image that is $224 \times 224$ pixels.  Next, the image is converted to an array, which is then resized to a 4D tensor.  In this case, since we are working with color images, each image has three channels.  Likewise, since we are processing a single image (or sample), the returned tensor will always have shape

$$
(1, 224, 224, 3).
$$

The `paths_to_tensor` function takes a numpy array of string-valued image paths as input and returns a 4D tensor with shape 

$$
(\text{nb_samples}, 224, 224, 3).
$$

Here, `nb_samples` is the number of samples, or number of images, in the supplied array of image paths.  It is best to think of `nb_samples` as the number of 3D tensors (where each 3D tensor corresponds to a different image) in your dataset!

```python
from keras.preprocessing import image                  
from tqdm import tqdm

def path_to_tensor(img_path):
    # loads RGB image as PIL.Image.Image type
    img = image.load_img(img_path, target_size=(224, 224))
    # convert PIL.Image.Image type to 3D tensor with shape (224, 224, 3)
    x = image.img_to_array(img)
    # convert 3D tensor to 4D tensor with shape (1, 224, 224, 3) and return 4D tensor
    return np.expand_dims(x, axis=0)

def paths_to_tensor(img_paths):
    list_of_tensors = [path_to_tensor(img_path) for img_path in tqdm(img_paths)]
    return np.vstack(list_of_tensors)
```

### Making Predictions with ResNet-50

Getting the 4D tensor ready for ResNet-50, and for any other pre-trained model in Keras, requires some additional processing.  First, the RGB image is converted to BGR by reordering the channels.  All pre-trained models have the additional normalization step that the mean pixel (expressed in RGB as $[103.939, 116.779, 123.68]$ and calculated from all pixels in all images in ImageNet) must be subtracted from every pixel in each image.  This is implemented in the imported function `preprocess_input`.  If you're curious, you can check the code for `preprocess_input` [here](https://github.com/fchollet/keras/blob/master/keras/applications/imagenet_utils.py).

Now that we have a way to format our image for supplying to ResNet-50, we are now ready to use the model to extract the predictions.  This is accomplished with the `predict` method, which returns an array whose $i$-th entry is the model's predicted probability that the image belongs to the $i$-th ImageNet category.  This is implemented in the `ResNet50_predict_labels` function below.

By taking the argmax of the predicted probability vector, we obtain an integer corresponding to the model's predicted object class, which we can identify with an object category through the use of this [dictionary](https://gist.github.com/yrevar/942d3a0ac09ec9e5eb3a). 

```python
from keras.applications.resnet50 import preprocess_input, decode_predictions

def ResNet50_predict_labels(img_path):
    # returns prediction vector for image located at img_path
    img = preprocess_input(path_to_tensor(img_path))
    return np.argmax(ResNet50_model.predict(img))
```

### Write a Dog Detector

While looking at the [dictionary](https://gist.github.com/yrevar/942d3a0ac09ec9e5eb3a), you will notice that the categories corresponding to dogs appear in an uninterrupted sequence and correspond to dictionary keys 151-268, inclusive, to include all categories from `'Chihuahua'` to `'Mexican hairless'`.  Thus, in order to check to see if an image is predicted to contain a dog by the pre-trained ResNet-50 model, we need only check if the `ResNet50_predict_labels` function above returns a value between 151 and 268 (inclusive).

We use these ideas to complete the `dog_detector` function below, which returns `True` if a dog is detected in an image (and `False` if not).

```python
### returns "True" if a dog is detected in the image stored at img_path
def dog_detector(img_path):
    prediction = ResNet50_predict_labels(img_path)
    return ((prediction <= 268) & (prediction >= 151)) 
```

__Answer:__ 

The model I trained closely resembles with hinted at model, with some minor changes for the input shape.

**First layer**

For the first layer in the network, we need to input the images, which are 224 pixels by 224 pixels. This convolutional layer will handle the most basic analysis of the images: deconstructing the images into edges, colors and other basic elements. Since the images are color, they will have a depth of 3 to account for each of the RGB values. This means the value of the input_shape parameter should be (224, 224, 3). The hinted-at model uses an input shape of 223, 223, 3, and it's unclear why the model would drop the one pixel. For subsequent convolutional layers, we do NOT specify an input shape, since the network isn't receiving additional inputs from outside the network.

The first convolutional layer needs to process information from every single pixel in the image. But our network is concerned with pattern recognition across groups of pixels, not the individual pixels themselves. This is where the filters and kernel size feature. The filters indicate the number of different patterns we're detecting within the image; At the first level, we're only collecting information on the building blocks of the more complex aspects of the images. This layer will detect patterns like edges that layers later on in the network will combine to capture things like nose shape, floppy or pointy ears, coat texture, and coat color. 

Next we consider the kernel size. A smaller kernel size can be faster *and* it helps safeguard against overfitting and memorizing the training data. This is because a smaller kernel size breaks the image into smaller pieces of information, the neural network analyzes the individual pieces rather than big chunks of the images. This model uses the smallest kernel size of 2, which means that each filter consists of a 2 x 2 square that passes over the image.

The default stride parameter of two is appropriate for the filter size we've selected. If the stride was larger, we would skip over pixels that haven't been analyzed by the filter. If the stride were smaller, we would be repeating information already gathered by the filter.

For the 'padding' parameter, the course videos recommend always using 'same' so that no information is lost at the edges of the image. We've already resized these images so that they are uniformly 244 by 244 pixels, and there's technically no need to provide a buffer: since our filter size is two and our stride is two, we actually don't risk losing any pixels as our filter passes over the image: 224 is evenly divisible by two.

For all convolutional layers, we've used a Relu function as the activation function. This function is recommended by the course to use as the activation function for convolutional layers in CNNs because it helps guard against the 'vanishing gradient problem'. This happens in, for example, the sigmoid and softmax activation functions which have gradients that are very, very small. Because they are so small, gradient descent doesn't function well, which makes backpropogation through the network challenging. 

*Subsequent layers*

After each convolutional layer, there is a MaxPooling layer. These layers help reduce the dimensionality within the CNN. Reducing the dimensionality decreases the volume of information that we're passing through the network. Without it, as we add more filters in each layer, the number of parameters would be enormous. Additionally, the network doesn't *need* some of the details. The course videos recommended using pool_size of two and the default stride of two here because this reduces the dimensions by half of what they were in the previous layer. As we progress through the CNN, this loss of spatial information is acceptable: previous layers will have gathered the information that they need to build patterns based on the relative positioning of the spatial information. The later layers in the network only need to know whether or not those patterns occured.

The network contains multiple sets of convolutional and pooling layers. This structure helps the network progress from its most fundamental observations (e.g. edges, colors) to more complex observations, with each set of layers learning using the information from previous layers.

*Second to last layer*

The second-to-last layer of the network is a Global Average Pooling layer. It bridges the individual features cumulatively identified in the previous layers and collects them into feature maps to bundle together collections of features that are used to identify the breeds identified in the last dense layer of our network. Without this layer, the final layer would only receive individual features (e.g. coat type, ear floppiness) without understanding how those features fit together to show images of different breeds of dogs.

*Last Layer*

The last layer of the CNN should be a Dense layer because this layer is the one actually handling the predictions using all of the information it has accumulated in the layers. It needs information from all of the patterns we've detected in order to make its decisions, so it needs to be a fully connected, dense layer. We pass it a value of 133 because that is the number of dog breeds that we are detecting.

The softmax function is the appropriate activation function to use in this layer because it takes a number of classes and returns the probability that the object being analyzed is a member of that class. This is perfect for our task, since we have 133 different dog breeds, and we need to predict which breed mostly likely is represented by the image.

Alternatively, if we were building the human or dog classifier, we would use a sigmoid function. It returns the probability that an object is a member of a specific class.

```python
from keras.layers import Conv2D, MaxPooling2D, GlobalAveragePooling2D
from keras.layers import Dropout, Flatten, Dense
from keras.models import Sequential

model = Sequential()

model.add(Conv2D(filters=16, kernel_size=2, padding='same', activation='relu', 
                        input_shape=(224, 224, 3)))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=32, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(Conv2D(filters=64, kernel_size=2, padding='same', activation='relu'))
model.add(MaxPooling2D(pool_size=2))
model.add(GlobalAveragePooling2D())
model.add(Dense(133, activation='softmax'))

model.summary()
```

__Answer:__ 

Using transfer learning means that we get to incorporate all of the training done on low-level features of other models to fast-forward our own training. The degree of similarity between our data and the data used to train the model we're transfering from determines whether we need additional layers that combine lower-level feature data into something meaningful for our model.

If we were transfering learning from CNN that had been trained on data that was very different from our dog breed data, then we might need to perform some additional interpretation. For example, if the previous CNN had been trained on a data set filled with all different sorts of images, then its lower layers would still perform valuable tasks such as edge detection, but the upper layers may not understand how those edges combine to make, for example, a face. In this case, however, the bottleneck features already been trained specifically on dog data. That means all our CNN needs to do is create combinations of the bottleneck features, which is done here using a global average pooling layer. Our final layer will use these feature maps to determine the probability that a specific breed is present in the image given presence of a specific group of features.

As with the previous model I trained, the last layer should be a dense layer because it is the model that handles the final predictions. Again, we give it a value of 133 because that corresponds to the number of breeds we are predicting, and we use a softmax function as its activation function so that we can predict the probabilities of individual classes.
